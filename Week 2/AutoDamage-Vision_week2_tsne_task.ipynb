```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from tensorflow.keras.datasets import fashion_mnist
```

```python
(X, y), (_, _) = fashion_mnist.load_data()
```

```python
X_flat = X.reshape(len(X), -1)
X_flat = X_flat / 255.0
```

```python
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_flat[:3000])
y_subset = y[:3000]
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    X_tsne[:, 0],
    X_tsne[:, 1],
    c=y_subset,
    cmap="tab10",
    s=5
)
plt.colorbar(scatter)
plt.title("t-SNE Visualization of Fashion-MNIST")
plt.show()
```

Do clusters correspond to labels?

Yes. Items like shoes, bags, and shirts form visually distinct clusters.

Why should t-SNE not be used for training models?

t-SNE distorts global geometry and is non-deterministic; it is only for visualization.
