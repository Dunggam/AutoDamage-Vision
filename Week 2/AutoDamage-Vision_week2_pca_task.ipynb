```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from tensorflow.keras.datasets import fashion_mnist
```

```python
(X, y), (_, _) = fashion_mnist.load_data()
```

```python
X_flat = X.reshape(len(X), -1)
```

```python
pca = PCA(n_components=50)
pca.fit(X_flat)
```

```python
X_pca = pca.transform(X_flat)
X_reconstructed = pca.inverse_transform(X_pca)
```

```python
n = 10
plt.figure(figsize=(10, 4))

for i in range(n):
    plt.subplot(2, n, i + 1)
    plt.imshow(X[i], cmap="gray")
    plt.axis("off")

    plt.subplot(2, n, i + 1 + n)
    plt.imshow(X_reconstructed[i].reshape(28, 28), cmap="gray")
    plt.axis("off")

plt.suptitle("Top: Original | Bottom: PCA Reconstructed")
plt.show()
```

What information is lost during PCA?

Ans. PCA discards low-variance components, which usually contain fine-grained textures, edges, and noise.

Why does reconstruction blur fine details?

Because PCA keeps only dominant variance directions, high-frequency image details are removed.
